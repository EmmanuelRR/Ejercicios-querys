```python
def query_chinook(query_text):  
  conexion=sqlite3.connect('chinook.db')
  cursor=conexion.cursor()
  cursor.execute(query_text)
  names=[description[0] for description in cursor.description]
  result=pd.DataFrame(cursor.fetchall(), columns=names)
  conexion.commit()
  conexion.close()
  return result 
```

```python
query_chinook('SELECT * FROM employees').head(3)
```

```python
query_chinook("SELECT * FROM employees where LastName= 'Adams'")
```

¿qué paises tiene mas faacturas? utilice la tabla facturas para determinar los paises que tienen mas facturas. proporcione una tabla de BillingCountry y Ivoices ordenados por el numero de facturas para cada pais. EL pais con mas facturas debe aparecer primero

```python
query= """SELECT BillingCountry, COUNT(*) AS num_fac 
FROM invoices
GROUP BY  BillingCountry
ORDER BY  num_fac DESC LIMIT 5"""
query_chinook(query)
```

¿Quién es el mejor cliente? El cliente que haya gastado más dinero será el mejor cliente 

```python
query="""SELECT SUM(Total) AS Total_fac, FirstName, LastName 
FROM (SELECT * FROM invoices LEFT JOIN customers
ON invoices.CustomerId=customers.CustomerId) 
GROUP BY CustomerId 
ORDER BY Total_fac DESC LIMIT 5"""
query_chinook(query)
```
